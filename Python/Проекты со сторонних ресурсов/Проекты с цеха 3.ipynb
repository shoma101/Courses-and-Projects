2.2 Работа с множеством событий

```python
events = data_first['event_name'].value_counts(normalize=True)
events[events > 0.01]
```

```python
#Сгруппируем события
def group_event(event_name):
  main_events = ['todo_list_item.connect_ig_account.done', 'todo_list_item.verify_email.done', 'todo_list_item.try_ig_automation.started',
                 'web.page_view.tutorial.show', 'todo_list_item.try_ig_automation.done', 'no session']
  event_name = str(event_name)
  if event_name in main_events:
    event_group = event_name
  elif 'system.event' in event_name:
    event_group = 'system event'
  elif event_name[:6] == 'mobile':
    event_group = 'mobile event'
  elif 'web.builder' in event_name:
    event_group = 'web builder event'
  elif 'web.onboarding' in event_name:
    event_group = 'onboarding event'
  elif 'web.interface' in event_name:
    event_group = 'web interface event'
  elif 'web.page' in event_name:
    event_group = 'web page event'
  elif 'web.dashboard' in event_name:
    event_group = 'web dashboard event'
  elif 'web.settings' in event_name:
    event_group = 'web settings event'
  elif 'web.home_tab' in event_name:
    event_group = 'web home tab event'
  else:
    event_group = 'other event'

  return event_group

# Обработаем строки с пропусками
data_first['ig_enable_dttm'] = pd.to_datetime(data_first['ig_enable_dttm'])
data_first = data_first.fillna({'event_name': 'no session', 'event_dttm': data_first['ig_enable_dttm']})

# И создадим новый признак "группа событий"
data_first['event_group'] = data_first['event_name'].apply(group_event)
```

2.3 Кластеризация пользователей на основе поведения в первой сессии

```python
# Зададим необходимые поля для работы с библиотекой retentioneering
retentioneering.config.update({'user_col': 'client_id',
                               'event_col': 'event_name', # будем использовать собственно события
                               'event_time_col': 'event_dttm'
                               })

# Step matrix для 20 первых событий первой сессии
data_first.rete.step_matrix(max_steps=20, thresh=0.025)
```

```python
# Граф поведения пользователей в течение первой сессии
data_first.rete.plot_graph(norm_type='full',
                           thresh = 0.01,
                           weight_col=None)
```

```python
# Визуализация кластеров
data_first.rete.get_clusters(method='kmeans',
                       feature_type='tfidf',
                       n_clusters=5,
                       ngram_range=(1,1),
                       plot_type='cluster_bar')
```

```python
# Граф поведения пользователей для определённого кластера
clus = data_first.rete.filter_cluster(4)
clus.rete.plot_graph(thresh=0.1,
                       weight_col='client_id')
```

```python
# Визуализация кластеров - каждая последовательность событий (вектор) преообразована в точку на плоскости

data_first.rete.project(plot_type ='clusters',
                  method='tsne')
```

```python
# Визуализация различий между кластерами
g1 = data_first.rete.cluster_mapping[0]
g2 = data_first.rete.cluster_mapping[1]

data_first.rete.step_matrix(max_steps=20,
                      thresh = 0.01,
                      groups=(g1, g2));
```

```python
# Выведем статистическую информацию по кластерам
clus_statistics = pd.DataFrame(columns=['cluster', 'duration min', 'events', 'total clients', 'total accounts', 'activated'])

# преобразуем в формат даты признак, с которым будем работать
data_first['ig_activation_dttm'] = pd.to_datetime(data_first['ig_activation_dttm'])
data_first['activation_date'] = data_first['ig_activation_dttm'].apply(lambda x: x.date())

# создадим бинарный признак, показывающий, произошла ли активация в первую сессию
#data_first['first_day_activation'] = data_first['account_date'] == data_first['activation_date']
#data_first['first_day_activation'] = data_first['first_day_activation'].apply(lambda x: 1 if x else 0)

for i in range (0, 5):
  clus = data_first.rete.cluster_mapping[i] # подмножество пользователей, относящихся к кластеру i

  # выбираем первые сессии для пользователей нашего кластера, фильтруем, группируем по номеру сессии
  filtered = data_first[data_first['client_id'].isin(clus)]

  # чтобы рассчитать длительность сессии
  grouped = filtered.groupby('session_id').agg(start=('event_dttm', 'min'), end=('event_dttm', 'max'), events=('event_name', 'count'),
                                              #activation=('first_day_activation', 'sum')
                                              )
  grouped['duration'] = grouped['end'] - grouped['start']
  #grouped['activation'] = grouped['activation'] / grouped['events'] # 1 в случае, если активация произошла в первой сессии

  new_row = {'cluster': i,
             'duration min': round(grouped['duration'].mean().total_seconds()/60, 2), # длительность сессии в минутах
             'events': round(grouped['events'].mean(), 2), # среднее кол-во событий в сессии (не уникальных)
             'total clients': filtered['client_id'].nunique(),  # кол-во клиентов
             'total accounts': filtered['account_id'].nunique(),  # кол-во ботов
             'activated': filtered['ig_activation_dttm'].nunique() # и кол-во активированных клиентов
             }
  clus_statistics = clus_statistics.append(new_row, ignore_index=True)

total_row = {'cluster': 'total', 'duration min': 13.52, 'events': 23.50, 'total clients': 11754, 'total accounts': 12018, 'activated': 3374} # информация из предыдущих расчётов
clus_statistics = clus_statistics.append(total_row, ignore_index=True)

# конверсия
clus_statistics['client act. rate'] = round(clus_statistics['activated']/clus_statistics['total clients'], 2)
clus_statistics['account act. rate'] = round(clus_statistics['activated']/clus_statistics['total accounts'], 2)
```
