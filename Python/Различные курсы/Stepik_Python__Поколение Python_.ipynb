```python
#–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é matrix(), –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç, –∑–∞–ø–æ–ª–Ω—è–µ—Ç –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞. –ü—Ä–∏ —ç—Ç–æ–º (–≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤) –æ–Ω–∞ –¥–æ–ª–∂–Ω–∞ –≤–µ—Å—Ç–∏ —Å–µ–±—è —Ç–∞–∫:
def matrix(n=1, m=None, value=0):
    if m is None:
        m = n
    return [[value] * m for _ in range(n)]


print(matrix())                   # –º–∞—Ç—Ä–∏—Ü–∞ 1 √ó 1 –∏–∑ 0
print(matrix(3))                  # –º–∞—Ç—Ä–∏—Ü–∞ 3 √ó 3 –∏–∑ 0
print(matrix(2, 5))               # –º–∞—Ç—Ä–∏—Ü–∞ 2 √ó 5 –∏–∑ 0
print(matrix(3, 4, 9))            # –º–∞—Ç—Ä–∏—Ü–∞ 3 √ó 4 –∏–∑ 9
```

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é sq_sum(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–ª–æ–≤—ã—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—É–º–º—É –∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤.

```python
def sq_sum(*args):
    summa=0
    for i in range(0, len(args)):
        summa+=args[i]**2
    return summa

```


–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é mean(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã—Ö –≤ –Ω–µ–µ —á–∏—Å–ª–æ–≤—ã—Ö (int –∏–ª–∏ float) –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤.


```python
def mean(*args):
    args = [x for x in args if type(x) == int or type(x) == float]

    return sum(args) / len(args) if args else 0

```

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é greet(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ —Å—Ç—Ä–æ–∫ –∏–º–µ–Ω (–∫–∞–∫ –º–∏–Ω–∏–º—É–º –æ–¥–Ω–æ) –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–∏–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –æ–±—Ä–∞–∑—Ü–æ–º

```python
def greet(name, *args):
    listochek=[]
    listochek.append(name)
    if args==None:
        return f"Hello, {' and '.join(listochek)}!"
    else:
        for i in args:
            listochek.append(i)
        return f"Hello, {' and '.join(listochek)}!"
```

```python
def greet(name, *args):
    return f'Hello, {" and ".join((name,) + args)}!'
```

```python
def print_products(*args):
    listok=[]
    for i in range(0, len(args)):
        if type(args[i]) is str and args[i] not in ("''"):
            listok.append(args[i])
    for j in range(0, len(listok)):
        print(f'{j+1}) {listok[j]}')
    if len(listok)==0:
        print('–ù–µ—Ç –ø—Ä–æ–¥—É–∫—Ç–æ–≤')

```

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é compose(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–µ –¥—Ä—É–≥–∏—Ö –æ–¥–Ω–æ–∞—Ä–≥—É–º–µ–Ω—Ç–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–∏ f –∏ g –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤—É—é —Ñ—É–Ω–∫—Ü–∏—é. –≠—Ç–∞ –Ω–æ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ç–∞–∫–∂–µ –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç x –∏ –ø—Ä–∏–º–µ–Ω—è—Ç—å –∫ –Ω–µ–º—É –∏—Å—Ö–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –Ω—É–∂–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ: –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–π f –∏ g –ø–æ—Ä—è–¥–æ–∫ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –¥–æ–ª–∂–µ–Ω –≤—ã–≥–ª—è–¥–µ—Ç—å, –∫–∞–∫ f(g(x)).

```python
def compose(f, g):
    return lambda x: f(g(x))

```

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é arithmetic_operation(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å–∏–º–≤–æ–ª –æ–¥–Ω–æ–π –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π (+, -, *, /) –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é –¥–≤—É—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–π –æ–ø–µ—Ä–∞—Ü–∏–∏.



```python
from operator import *
kw = {"+" : add,
       "-" : sub,
       "*" : mul,
       "/" : truediv}
def arithmetic_operation(operation):
    def schet(x,y):
        return kw[operation](x,y)
    return schet
```

–î–ª–∏–Ω–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏
–í–∞–º –¥–æ—Å—Ç—É–ø–µ–Ω —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª lines.txt, –≤ –∫–æ—Ç–æ—Ä–æ–º –∑–∞–ø–∏—Å–∞–Ω—ã —Å—Ç—Ä–æ–∫–∏ —Ç–µ–∫—Å—Ç–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –Ω–∞–∏–±–æ–ª—å—à–µ–π –¥–ª–∏–Ω—ã –∏–∑ —Ñ–∞–π–ª–∞, –Ω–µ –º–µ–Ω—è—è –∏—Ö –ø–æ—Ä—è–¥–æ–∫.

```python
with open('lines.txt', 'r', encoding='utf-8') as file:
    files =list(map(str.strip, file.readlines()))
    leni=len(max(files, key=len))
    print(*list(filter(lambda x:leni==len(x),files)), sep='\n')
```

–í–∞–º –¥–æ—Å—Ç—É–ø–µ–Ω —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª numbers.txt, –∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∫–æ—Ç–æ—Ä–æ–≥–æ –º–æ–∂–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å –æ–¥–Ω–æ –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã—Ö –æ–¥–Ω–∏–º –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ø—Ä–æ–±–µ–ª–∞–º–∏.

```python
with open('numbers.txt', 'r', encoding='utf-8') as file:
    for line in file:
        k = list(map(int,line.split()))
        print(sum(k))
```

–°—É–º–º–∞ —á–∏—Å–µ–ª –≤ —Ñ–∞–π–ª–µ
–í–∞–º –¥–æ—Å—Ç—É–ø–µ–Ω —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª nums.txt. –í —Ñ–∞–π–ª–µ –º–æ–≥—É—Ç –±—ã—Ç—å –∑–∞–ø–∏—Å–∞–Ω—ã —Ü–µ–ª—ã–µ –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –∏ –≤—Å–µ, —á—Ç–æ —É–≥–æ–¥–Ω–æ. –ß–∏—Å–ª–æ–º –Ω–∞–∑–æ–≤–µ–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–π –∏ –±–æ–ª–µ–µ —Ü–∏—Ñ—Ä, –∏–¥—É—â–∏—Ö –ø–æ–¥—Ä—è–¥ (—á–∏—Å–ª–æ –≤—Å–µ–≥–¥–∞ –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ).

```python
with open('nums.txt', encoding='utf-8') as file:
    mkey=[[i.replace(i, ' ') if i.isdigit()==False else i for i in line] for line in file]
    k = [''.join(l).strip().split() for l in mkey]
    print(sum([sum([int(i) for i in j]) for j in k]))

```

```python
with open('file.txt', encoding='utf-8') as file:
    k=[i.strip() for i in file]
    count=0
    counti=0
    for i in range(0, len(k)):
        for j in k[i].split():
            counti+=1
            for r in j:
                if r.isalpha():
                    count+=1

print(f"Input file contains:\n{count} letters\n{counti} words\n{i+1 if i!=0 else 0} lines")

```

–í–∞–º –¥–æ—Å—Ç—É–ø–µ–Ω —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª goats.txt –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–∞–ø–∏—Å–∞–Ω–æ —Å–ª–æ–≤–æ COLOURS, –¥–∞–ª–µ–µ –∏–¥–µ—Ç —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ü–≤–µ—Ç–æ–≤ –∫–æ–∑–ª–æ–≤. –ó–∞—Ç–µ–º –∏–¥–µ—Ç —Å—Ç—Ä–æ–∫–∞ —Å–æ —Å–ª–æ–≤–æ–º GOATS, –∏ –¥–∞–ª–µ–µ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ–∑–ª–æ–≤ —Ä–∞–∑–Ω—ã—Ö —Ü–≤–µ—Ç–æ–≤. –ü–µ—Ä–µ—á–µ–Ω—å –∫–æ–∑–ª–æ–≤ –≤–∫–ª—é—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫–∏ –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —Å–ø–∏—Å–∫–∞.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É —Å–æ–∑–¥–∞–Ω–∏—è —Ñ–∞–π–ª–∞ answer.txt –∏ –≤—ã–≤–æ–¥–∞ –≤ –Ω–µ–≥–æ —Å–ø–∏—Å–∫–∞ –∫–æ–∑–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—Ç —É—Å–ª–æ–≤–∏—é –∑–∞–≥–∞–¥–∫–∏ –æ—Ç –ñ–∞–∫–∞ –§—Ä–µ—Å–∫–æ.

```python
with open('goats.txt', 'r', encoding='utf-8') as goat, open('answer.txt', 'w', encoding='utf-8') as answer:
    colours, goats_colour =[], []
    file=goat.readlines()
    for line in file:
        if line.strip() !='GOATS':
            colours.append(line.strip())
        else:
            break
    colours.remove('COLOURS')
    for line in file[::-1]:
        if line.strip() !='GOATS':
            goats_colour.append(line.strip())
        else:
            break
    result = {}
    for goatss in goats_colour:
        if goatss in colours:
            result[goatss] = result.get(goatss, 0) + 1

    for i, dict in sorted(result.items()):
        if int(dict/len(goats_colour)*100)>7:
            print(i, file=answer)
```

```python
with open('goats.txt') as f1, open('answer.txt', 'w') as f2:
    cont = f1.read().split('\n')
    colors, goats = cont[1:cont.index('GOATS')], cont[cont.index('GOATS')+1:]
    print(*sorted(filter(lambda x: goats.count(x) / len(goats) > 0.07, colors)), sep='\n', file=f2)
```

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞ —Å –∏–º–µ–Ω–µ–º —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã–≤–æ–¥—è—â—É—é –Ω–∞ —ç–∫—Ä–∞–Ω —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —ç—Ç–æ–≥–æ —Ñ–∞–π–ª–∞, –Ω–æ —Å –∑–∞–º–µ–Ω–æ–π –≤—Å–µ—Ö –∑–∞–ø—Ä–µ—â–µ–Ω–Ω—ã—Ö —Å–ª–æ–≤ –∑–≤–µ–∑–¥–æ—á–∫–∞–º–∏ * (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–≤–µ–∑–¥–æ—á–µ–∫ —Ä–∞–≤–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –±—É–∫–≤ –≤ —Å–ª–æ–≤–µ).

–ó–∞–ø—Ä–µ—â–µ–Ω–Ω—ã–µ —Å–ª–æ–≤–∞, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞, —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ñ–∞–π–ª–µ forbidden_words.txt. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –≤—Å–µ —Å–ª–æ–≤–∞ –≤ —ç—Ç–æ–º —Ñ–∞–π–ª–µ –∑–∞–ø–∏—Å–∞–Ω—ã –≤ –Ω–∏–∂–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ.

```python
with open("forbidden_words.txt", 'r', encoding='utf-8') as forbidden, open(input(), 'r', encoding='utf-8') as checki:
    files = list(map(lambda x: x, forbidden.read().split()))
    checkii = checki.read()
    mini=checkii.lower()
    for i in files:
        mini=mini.replace(i.lower(), (len(i))*'*')
    for i in range(0, len(mini)):
        if mini[i]=='*':
            print(mini[i], end='')
        else:
            print(checkii[i], end='')
```

–í–∞–º –¥–æ—Å—Ç—É–ø–µ–Ω —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª cyrillic.txt, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π —Ç–µ–∫—Å—Ç. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è —Ç—Ä–∞–Ω—Å–ª–∏—Ç–µ—Ä–∞—Ü–∏–∏ —ç—Ç–æ–≥–æ —Ñ–∞–π–ª–∞, —Ç–æ –µ—Å—Ç—å –∑–∞–º–µ–Ω—ã –∫–∏—Ä–∏–ª–ª–∏—á–µ—Å–∫–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤ –Ω–∞ –ª–∞—Ç–∏–Ω—Å–∫–∏–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ–π. –í—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã –Ω–∞–¥–æ –æ—Å—Ç–∞–≤–∏—Ç—å –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π. –†–µ–∑—É–ª—å—Ç–∞—Ç —Ç—Ä–∞–Ω—Å–ª–∏—Ç–µ—Ä–∞—Ü–∏–∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è –∑–∞–ø–∏—Å–∞—Ç—å –≤ —Ñ–∞–π–ª transliteration.txt.

```python
d = {
    '–∞': 'a', '–∫': 'k', '—Ö': 'h', '–±': 'b', '–ª': 'l', '—Ü': 'c', '–≤': 'v', '–º': 'm', '—á': 'ch',
    '–≥': 'g', '–Ω': 'n', '—à': 'sh', '–¥': 'd', '–æ': 'o', '—â': 'shh', '–µ': 'e', '–ø': 'p', '—ä': '*',
    '—ë': 'jo', '—Ä': 'r', '—ã': 'y', '–∂': 'zh', '—Å': 's', '—å': "'", '–∑': 'z', '—Ç': 't', '—ç': 'je',
    '–∏': 'i', '—É': 'u', '—é': 'ju', '–π': 'j', '—Ñ': 'f', '—è': 'ya', '–ê': 'A', '–ö': 'K', '–•': 'H', '–ë': 'B', '–õ': 'L',
    '–¶': 'C', '–í': 'V', '–ú': 'M', '–ß': 'Ch',
    '–ì': 'G', '–ù': 'N', '–®': 'Sh', '–î': 'D', '–û': 'O', '–©': 'Shh', '–ï': 'E', '–ü': 'P', '–™': '*',
    '–Å': 'Jo', '–†': 'R', '–´': 'Y', '–ñ': 'Zh', '–°': 'S', '–¨': "'", '–ó': 'Z', '–¢': 'T', '–≠': 'Je',
    '–ò': 'I', '–£': 'U', '–Æ': 'Ju', '–ô': 'J', '–§': 'F', '–Ø': 'Ya'
}
with open('cyrillic.txt', 'r', encoding='utf-8') as curil, open('transliteration.txt', 'w', encoding='utf-8') as outik:
    file=curil.readlines()
    for i in file:
        for j in i:
            if j in d:
                print(d[j], end='', file=outik)
            else:
                print(j, end='', file=outik)
```

```python
# –ë–µ—Ä–µ–º —Å–ª–æ–≤–∞—Ä—å –î–∞—Ä—å–∏ –ë–æ—Ä–∑–æ–≤–æ–π
d = {
    '–∞': 'a', '–∫': 'k', '—Ö': 'h', '–±': 'b', '–ª': 'l', '—Ü': 'c', '–≤': 'v', '–º': 'm', '—á': 'ch',
    '–≥': 'g', '–Ω': 'n', '—à': 'sh', '–¥': 'd', '–æ': 'o', '—â': 'shh', '–µ': 'e', '–ø': 'p', '—ä': '*',
    '—ë': 'jo', '—Ä': 'r', '—ã': 'y', '–∂': 'zh', '—Å': 's', '—å': "'", '–∑': 'z', '—Ç': 't', '—ç': 'je',
    '–∏': 'i', '—É': 'u', '—é': 'ju', '–π': 'j', '—Ñ': 'f', '—è': 'ya'
    }

# –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–≥–ª–∞–≤–Ω—ã–µ –±—É–∫–≤—ã
d.update({k.upper(): (v[0].upper() + v[1:]) for k, v in d.items()})

# –í—É–∞–ª—è!
with open('cyrillic.txt', 'r', encoding='utf-8') as inp, open('transliteration.txt', 'w', encoding='utf-8') as out:
    text= inp.read()
    print(''.join(map(lambda ch: d.get(ch, ch), text)), file=out)
```

–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∫–æ–º–º–µ–Ω—Ç—ã üå∂Ô∏è
–ü—Ä–∏ –Ω–∞–ø–∏—Å–∞–Ω–∏–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –≤ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –æ–ø–∏—Å—ã–≤–∞—Ç—å –Ω–∞–∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏, –µ–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. –ß–∞—Å—Ç–æ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç—ã –æ—Ç–∫–ª–∞–¥—ã–≤–∞—é—Ç –Ω–∞–ø–∏—Å–∞–Ω–∏–µ —Ç–∞–∫–∏—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ –Ω–∞–ø–æ—Å–ª–µ–¥–æ–∫, –∞ –ø–æ—Ç–æ–º –∏ –≤–æ–≤—Å–µ –∑–∞–±—ã–≤–∞—é—Ç –æ –Ω–∏—Ö üòÇ.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞ —Å –∏–º–µ–Ω–µ–º —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞–ø–∏—Å–∞–Ω –∫–æ–¥ –Ω–∞ —è–∑—ã–∫–µ Python. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã–≤–æ–¥—è—â—É—é –Ω–∞ —ç–∫—Ä–∞–Ω –∏–º–µ–Ω–∞ –≤—Å–µ—Ö —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –ø–æ—è—Å–Ω—è—é—â–∏–π –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π. –ë—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –ª—é–±–∞—è —Å—Ç—Ä–æ–∫–∞, –Ω–∞—á–∏–Ω–∞—é—â–∞—è—Å—è —Å–æ —Å–ª–æ–≤–∞ def –∏ –ø—Ä–æ–±–µ–ª–∞, —è–≤–ª—è–µ—Ç—Å—è –Ω–∞—á–∞–ª–æ–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏. –§—É–Ω–∫—Ü–∏—è —Å–æ–¥–µ—Ä–∂–∏—Ç –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π, –µ—Å–ª–∏ –ø–µ—Ä–≤—ã–π —Å–∏–º–≤–æ–ª –ø—Ä–µ–¥—ã–¥—É—â–µ–π —Å—Ç—Ä–æ–∫–∏ - #.

```python
with open(input(), 'r', encoding='utf-8') as forbidden:
    jj = forbidden.readlines()
    listok = []
    for i in range(0, len(jj)):
        if 'def' in jj[i] and '#' not in jj[i - 1]:
            listok.append(jj[i])
    listochek = []
    for i in listok:
        listochek.append(i.strip()[i.index(' '):i.index('(')].strip())
if len(listochek)>0:
    for i in listochek:
        print(i, sep='')
else:
    print('Best Programming Team')
```
