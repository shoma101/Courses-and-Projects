```python
#Напишите функцию matrix(), которая создает, заполняет и возвращает матрицу заданного размера. При этом (в зависимости от переданных аргументов) она должна вести себя так:
def matrix(n=1, m=None, value=0):
    if m is None:
        m = n
    return [[value] * m for _ in range(n)]


print(matrix())                   # матрица 1 × 1 из 0
print(matrix(3))                  # матрица 3 × 3 из 0
print(matrix(2, 5))               # матрица 2 × 5 из 0
print(matrix(3, 4, 9))            # матрица 3 × 4 из 9
```

Напишите функцию sq_sum(), которая принимает произвольное количество числовых аргументов и возвращает сумму их квадратов.

```python
def mean(*args):
    args = [x for x in args if type(x) == int or type(x) == float]

    return sum(args) / len(args) if args else 0

```

Если среди переданных аргументов нет ни одного продукта, необходимо вывести текст Нет продуктов.



```python
def print_products(*args):
    listok=[]
    for i in range(0, len(args)):
        if type(args[i]) is str and args[i] not in ("''"):
            listok.append(args[i])
    for j in range(0, len(listok)):
        print(f'{j+1}) {listok[j]}')
    if len(listok)==0:
        print('Нет продуктов')

```

```python
def print_products(*args):
    cnt = 0
    for e in args:
        if type(e) == str and e:
            cnt += 1
            print(f'{cnt}) {e}')
    if cnt == 0:
        print('Нет продуктов')
```

Напишите функцию compose(), которая принимает на вход две других одноаргументных функции f и g и возвращает новую функцию. Эта новая функция также должна принимать один аргумент x и применять к нему исходные функции в нужном порядке: для функций f и g порядок применения должен выглядеть, как f(g(x)).

Напишите функцию arithmetic_operation(), которая принимает символ одной из четырех арифметических операций (+, -, *, /) и возвращает функцию двух аргументов для соответствующей операции.



```python
from operator import *
kw = {"+" : add,
       "-" : sub,
       "*" : mul,
       "/" : truediv}
def arithmetic_operation(operation):
    def schet(x,y):
        return kw[operation](x,y)
    return schet
```

Вам доступен текстовый файл goats.txt в первой строке которого написано слово COLOURS, далее идет список всех возможных цветов козлов. Затем идет строка со словом GOATS, и далее непосредственно перечисление козлов разных цветов. Перечень козлов включает только строки из первого списка.

Напишите программу создания файла answer.txt и вывода в него списка козлов, которые удовлетворяют условию загадки от Жака Фреско.

```python
with open('goats.txt', 'r', encoding='utf-8') as goat, open('answer.txt', 'w', encoding='utf-8') as answer:
    colours, goats_colour =[], []
    file=goat.readlines()
    for line in file:
        if line.strip() !='GOATS':
            colours.append(line.strip())
        else:
            break
    colours.remove('COLOURS')
    for line in file[::-1]:
        if line.strip() !='GOATS':
            goats_colour.append(line.strip())
        else:
            break
    result = {}
    for goatss in goats_colour:
        if goatss in colours:
            result[goatss] = result.get(goatss, 0) + 1

    for i, dict in sorted(result.items()):
        if int(dict/len(goats_colour)*100)>7:
            print(i, file=answer)
```

```python
with open('goats.txt') as f1, open('answer.txt', 'w') as f2:
    cont = f1.read().split('\n')
    colors, goats = cont[1:cont.index('GOATS')], cont[cont.index('GOATS')+1:]
    print(*sorted(filter(lambda x: goats.count(x) / len(goats) > 0.07, colors)), sep='\n', file=f2)
```

Запрещенные слова, разделенные символом пробела, хранятся в текстовом файле forbidden_words.txt. Гарантируется, что все слова в этом файле записаны в нижнем регистре.

```python
with open("forbidden_words.txt", 'r', encoding='utf-8') as forbidden, open(input(), 'r', encoding='utf-8') as checki:
    files = list(map(lambda x: x, forbidden.read().split()))
    checkii = checki.read()
    mini=checkii.lower()
    for i in files:
        mini=mini.replace(i.lower(), (len(i))*'*')
    for i in range(0, len(mini)):
        if mini[i]=='*':
            print(mini[i], end='')
        else:
            print(checkii[i], end='')
```

Транслитерация 🌶️
Транслитерация — передача знаков одной письменности знаками другой письменности, при которой каждый знак (или последовательность знаков) одной системы письма передаётся соответствующим знаком (или последовательностью знаков) другой системы письма.

Вам доступен текстовый файл cyrillic.txt, содержащий текст. Напишите программу для транслитерации этого файла, то есть замены кириллических символов на латинские в соответствии с предложенной таблицей. Все остальные символы надо оставить без изменений. Результат транслитерации требуется записать в файл transliteration.txt.

```python
d = {
    'а': 'a', 'к': 'k', 'х': 'h', 'б': 'b', 'л': 'l', 'ц': 'c', 'в': 'v', 'м': 'm', 'ч': 'ch',
    'г': 'g', 'н': 'n', 'ш': 'sh', 'д': 'd', 'о': 'o', 'щ': 'shh', 'е': 'e', 'п': 'p', 'ъ': '*',
    'ё': 'jo', 'р': 'r', 'ы': 'y', 'ж': 'zh', 'с': 's', 'ь': "'", 'з': 'z', 'т': 't', 'э': 'je',
    'и': 'i', 'у': 'u', 'ю': 'ju', 'й': 'j', 'ф': 'f', 'я': 'ya', 'А': 'A', 'К': 'K', 'Х': 'H', 'Б': 'B', 'Л': 'L',
    'Ц': 'C', 'В': 'V', 'М': 'M', 'Ч': 'Ch',
    'Г': 'G', 'Н': 'N', 'Ш': 'Sh', 'Д': 'D', 'О': 'O', 'Щ': 'Shh', 'Е': 'E', 'П': 'P', 'Ъ': '*',
    'Ё': 'Jo', 'Р': 'R', 'Ы': 'Y', 'Ж': 'Zh', 'С': 'S', 'Ь': "'", 'З': 'Z', 'Т': 'T', 'Э': 'Je',
    'И': 'I', 'У': 'U', 'Ю': 'Ju', 'Й': 'J', 'Ф': 'F', 'Я': 'Ya'
}
with open('cyrillic.txt', 'r', encoding='utf-8') as curil, open('transliteration.txt', 'w', encoding='utf-8') as outik:
    file=curil.readlines()
    for i in file:
        for j in i:
            if j in d:
                print(d[j], end='', file=outik)
            else:
                print(j, end='', file=outik)
```

```python
# Берем словарь Дарьи Борзовой
d = {
    'а': 'a', 'к': 'k', 'х': 'h', 'б': 'b', 'л': 'l', 'ц': 'c', 'в': 'v', 'м': 'm', 'ч': 'ch',
    'г': 'g', 'н': 'n', 'ш': 'sh', 'д': 'd', 'о': 'o', 'щ': 'shh', 'е': 'e', 'п': 'p', 'ъ': '*',
    'ё': 'jo', 'р': 'r', 'ы': 'y', 'ж': 'zh', 'с': 's', 'ь': "'", 'з': 'z', 'т': 't', 'э': 'je',
    'и': 'i', 'у': 'u', 'ю': 'ju', 'й': 'j', 'ф': 'f', 'я': 'ya'
    }

# Добавляем заглавные буквы
d.update({k.upper(): (v[0].upper() + v[1:]) for k, v in d.items()})

# Вуаля!
with open('cyrillic.txt', 'r', encoding='utf-8') as inp, open('transliteration.txt', 'w', encoding='utf-8') as out:
    text= inp.read()
    print(''.join(map(lambda ch: d.get(ch, ch), text)), file=out)
```

Пропущенные комменты 🌶️
При написании собственных функций рекомендуется в комментарии описывать назначение функции, ее параметры и возвращаемое значение. Часто программисты откладывают написание таких комментариев напоследок, а потом и вовсе забывают о них 😂.

На вход программе подается строка текста с именем текстового файла, в котором написан код на языке Python. Напишите программу, выводящую на экран имена всех функций для которых отсутствует поясняющий комментарий. Будем считать, что любая строка, начинающаяся со слова def и пробела, является началом определения функции. Функция содержит комментарий, если первый символ предыдущей строки - #.

```python
with open(input(), 'r', encoding='utf-8') as forbidden:
    jj = forbidden.readlines()
    listok = []
    for i in range(0, len(jj)):
        if 'def' in jj[i] and '#' not in jj[i - 1]:
            listok.append(jj[i])
    listochek = []
    for i in listok:
        listochek.append(i.strip()[i.index(' '):i.index('(')].strip())
if len(listochek)>0:
    for i in listochek:
        print(i, sep='')
else:
    print('Best Programming Team')
```
