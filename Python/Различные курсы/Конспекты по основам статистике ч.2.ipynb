```python
import random
from math import sqrt
def distance_chi():
  eagles = 0
  tails = 0
  for i in range(60):
    if random.randint(0, 1):
      eagles+=1
    else:
      tails += 1
  return ((eagles-30)**2/30 +(tails-30)**2/30)
listok=[]
for i in range(10000):
  listok.append(distance_chi())
```

```python
import seaborn as sns
sns.distplot(listok)
```

```python
# посмотрим еще раз на значение, которое мы получали в нашем
# конкретном наблюдении с выпаданием орла и решки (когда получили 40 орлов!):
exp_eagles = 40
exp_tails = 20
# расчитываем расстояние хи-квадрат Пирсона:
# ((O1 - E1)|sqrt(E1))^2 + ... + ((On - En)|sqrt(En))^2
res_of_exp_chi_2 = (exp_eagles-30)**2/30 + (exp_tails-30)**2/30
print(res_of_exp_chi_2)
```

```python
count_of_result=0
for i in listok:
  if i>= res_of_exp_chi_2:
    count_of_result+=1
print(count_of_result)
```

```python
from scipy.stats import bartlett
from scipy.stats import shapiro
from scipy.stats import kruskal
from scipy.stats import f_oneway
```

```python
data = pd.read_csv(r'https://stepik.org/media/attachments/course/524/s_anova_test.csv')
```

```python
data
```

```python
def smart_anova(data):
  group_names=data.groupby(['y']).mean().index
  all_samples = [data.x[data.y ==group_name] for group_name in group_names]
  t_statistics, p_bartlett = bartlett(*all_samples)
  if p_bartlett < 0.05:
    return {'KW': kruskal(*all_samples).pvalue}
  p_shapiro_values = [shapiro(data.x[data.y == group_name].values).pvalue for group_name in group_names]
  for p in p_shapiro_values:
      if p < 0.05:
          return {'KW' : kruskal(*all_samples).pvalue}
    # ИНАЧЕ - при помощи дисперсионного анализа One-way ANOVA
  return {'ANOVA' : [f_oneway(*all_samples).pvalue]}

smart_anova(data)
```

```python
def get_pca2(df, threshold=0.9):
  pca = PCA(n_components = df.shape[1])
  fit = pca.fit_transform(df)
  n_components = sum(np.cumsum(pca.explained_variance_ratio_) <threshold) +1
  return pd.concat([df, pd.DataFrame(fit[:, n_components])], axis= 1)
```
